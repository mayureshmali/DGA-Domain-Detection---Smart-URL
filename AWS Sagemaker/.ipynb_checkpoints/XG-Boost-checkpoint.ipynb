```python
import tldextract

def domain_extract(uri):
    ext = tldextract.extract(uri)
    if (not ext.suffix):
        return np.nan
    else:
        return ext.domain
```

```python
data_ingested_copy['domain'] = [ domain_extract(DNS_Address) for DNS_Address in data_ingested_copy['DNS_Address']]
```

```python
# Add a length field for the domain
data_ingested_copy['length'] = [len(x) for x in data_ingested_copy['domain']]
```

```python
# Calculating character entropy
import math
from collections import Counter
 
def entropy(s):
    p, lns = Counter(s), float(len(s))
    return -sum( count/lns * math.log(count/lns, 2) for count in p.values())
```

```python
# calculate vowels
def calc_vowels(y):
    num_vowel = 0
    vowels = list('aeiou')
    for char in y:
        if char in vowels:
            num_vowel += 1

    return num_vowel
```

```python
#Calculate number of digits
def calc_digits(z):
    num_digit = 0
    digits = list('0123456789')
    for char in z:
        if char in digits:
            num_digit += 1

    return num_digit
```

```python
# Maximum length of Consecutive consonants 
def consecutive_consonants(string):
    from itertools import groupby
    is_vowel = lambda char: char in "aAeEiIoOuU"
    best = 0
    listnames = ["".join(g) for v, g in groupby(string, key=is_vowel) if not v]
    for index in range(len(listnames)):
        if len(listnames[index]) > best:
            best = len(listnames[index])
    return best
```

```python
# Add a entropy field for the domain
data_ingested_copy['entropy'] = [entropy(x) for x in data_ingested_copy['domain']]
```

```python
# Add a vowels field for the domain
data_ingested_copy['vowels'] = [calc_vowels(x) for x in data_ingested_copy['domain']]
```

```python
# Add a digits field for the domain
data_ingested_copy['digits'] = [calc_digits(x) for x in data_ingested_copy['domain']]
```

```python
# Add a consec_consonants field for the domain
data_ingested_copy['consec_consonants'] = [consecutive_consonants(x) for x in data_ingested_copy['domain']]
```

```python
from sagemaker.amazon.amazon_estimator import get_image_uri

sess = sagemaker.Session()

container = get_image_uri(region, 'xgboost')

xgb = sagemaker.estimator.Estimator(container,
                                    role, 
                                    train_instance_count=1, 
                                    train_instance_type='ml.m4.xlarge',
                                    output_path='s3://{}/{}/output'.format(bucket, prefix),
                                    sagemaker_session=sess)

```

```python
xgb.set_hyperparameters(max_depth=2,
                        eta=0.2,
                        gamma=4,
                        min_child_weight=6,
                        subsample=0.8,
                        silent=0,
                        objective='binary:logistic',
                        num_round=100,
                        eval_metric='auc')

xgb.fit({'train': s3_input_train, 'validation': s3_input_validation})
```

```python


xgb_predictor = xgb.deploy(initial_instance_count=1,instance_type='ml.t2.medium')

```
